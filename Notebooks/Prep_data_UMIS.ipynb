```python
import pandas as pd
```

```python
from datetime import datetime

clinical_data_df = pd.read_csv('Clinical Metadata FDG PET_CT Lesions (1).csv')
#clinical_data_df['Study Date'] = clinical_data_df['Study Date'].apply(lambda x: x.replace("/", ""))

#Â add zeros to date so it is compatible w/ other data
clinical_data_df['Study Date'] = clinical_data_df['Study Date'].apply(lambda x: (datetime.strptime(x, "%m/%d/%Y")).strftime("%m%d%Y"))

clinical_data_df['filename'] = clinical_data_df['Subject ID'].astype(str) + "_" + clinical_data_df['Study Date'].astype(str)
```

```python
clinical_data_df
```

```python
train_files= 'UMIS/train_imgs.txt'
val_files = 'UMIS/val_imgs.txt'
test_files = 'UMIS/test_imgs.txt'

with open(train_files, 'r') as f:
    train_filenames = [line.strip() for line in f.readlines()]

with open(val_files, 'r') as f:
    val_filenames = [line.strip() for line in f.readlines()]

with open(test_files, 'r') as f:
    test_filenames = [line.strip() for line in f.readlines()]

# Filter filenames based on 'diagnosis' column
pos_train_filenames = [filename for filename in train_filenames if clinical_data_df.loc[clinical_data_df['filename'] == filename, 'diagnosis'].values[0] != 'NEGATIVE']

pos_val_filenames = [filename for filename in val_filenames if clinical_data_df.loc[clinical_data_df['filename'] == filename, 'diagnosis'].values[0] != 'NEGATIVE']

pos_test_filenames = [filename for filename in test_filenames if clinical_data_df.loc[clinical_data_df['filename'] == filename, 'diagnosis'].values[0] != 'NEGATIVE']

# Write the filtered filenames to new text files
with open('pos_train_imgs.txt', 'w') as f:
    f.write('\n'.join(pos_train_filenames))

with open('pos_val_imgs.txt', 'w') as f:
    f.write('\n'.join(pos_val_filenames))

with open('pos_test_imgs.txt', 'w') as f:
    f.write('\n'.join(pos_test_filenames))
```
